```python
# Dependencies and Setup
import csv
import pandas as pd
import os
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress 
import seaborn as sns 
from scipy import stats
```

```python
homelessdata = "../Data/PIT_AllYearsData_Output.csv"
homeless_df = pd.read_csv(homelessdata)
homeless_df
```

```python
homeless_df2 = homeless_df.pivot(index='State', columns='Year', values='Overall Homeless')
homeless_df2.head(20)
```

```python
homeless_df3 = pd.DataFrame(homeless_df2.iloc[:,0])
homeless_df3['2019'] = homeless_df2.iloc[:,4]
homeless_df3.head()
```

Interesting to note that even though the overall number of homeless slightly increases, the majority of states see a decrease in homeless population. This suggests a concentration over time of homeless towards a few states.

```python
multi_plot = homeless_df3.plot(kind="bar", figsize=(20,15), width=1)
multi_plot.set_xticklabels(homeless_df2.index, rotation=60)
plt.show()
```

```python
homeless_df2['Percent Change 15-19'] = (homeless_df2.iloc[:,4] - homeless_df2.iloc[:,0]) /homeless_df2.iloc[:,0] * 100

x_axis = np.arange(len(homeless_df2))
tick_locations = [value for value in x_axis]
plt.figure(figsize=(20,3))
plt.bar(x_axis, homeless_df2.iloc[:,5], color='r', alpha=0.5, align="center")
plt.xticks(tick_locations, homeless_df2.index, rotation="vertical")
plt.show()
```
